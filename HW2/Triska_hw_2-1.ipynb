# Data integration (Ukrainian Vehicles Data Set)

```python
import pandas as pd
import numpy as np
import json
```

### Read the data

```python
df = pd.read_csv("data/tz_opendata_z01012017_po31122017.csv", sep=';')
df.head()
```

### Remove duplicates

```python
df.drop_duplicates()
```

### Remove all quotes

```python
for i, col in enumerate(df.columns):
    if not np.issubdtype(df[col].dtype, np.number):
        df.iloc[:, i] = df.iloc[:, i].str.replace('"', '')
```

### Remove whitespaces and tabs

```python
df = df.apply(lambda x: x.str.strip() if x.dtype == "object" else x)
```

```python
df.head()
```


### Save a dictionary of unique operation codes and its meanings in external fle (op_codes.txt). Remove the redundant column oper_name.

```python
oper_name_dict = dict()

for item in df['oper_name']:
    res = item.split(' - ')
    oper_name_dict[res[0]] = res[1]

with open('op_codes.txt', 'w') as file:
     file.write(json.dumps(oper_name_dict))
        
del df['oper_name']
```

### We can also remove the column *oper_code*, since it's redundant without *oper_name*

```python
del df['oper_code']
```

### Translate 3 attributes’ values to english (color, kind, purpose).

```python
df.head()
```

```python
colors_eng = ['GREEN', 'RED', 'WHITE', 'GRAY', 'YELLOW', 'BLUE', 'ORANGE', 'BEIGE', 'BLACK', 'BROWN', 'PURPLE', 'UNKNOWN']
kinds_eng = ['LIGHT', 'TRAILER', 'SEMI-TRAILER', 'CARGO', 'MOPED', 'MOTORCYCLE', 'BUS', 'ATV', 'TRICYCLE', 'SPECIALISED']
purpose_eng = ['GENERAL', 'SPECIALISED', 'SPECIAL']

color_dict = {df["color"].unique()[i]: colors_eng[i] for i in range(len(colors_eng))}
kind_dict = {df["kind"].unique()[i]: kinds_eng[i] for i in range(len(kinds_eng))}
purpose_dict = {df["purpose"].unique()[i]: purpose_eng[i] for i in range(len(purpose_eng))}
```

```python
df["color"] = df["color"].apply(lambda x: color_dict[x])
df["kind"] = df["kind"].apply(lambda x: kind_dict[x])
df["purpose"] = df["purpose"].apply(lambda x: purpose_dict[x])
```

```python
df.head()
```

### The name of the *d_reg* column is not clear enough in my opinion, so I will change it to *reg_date*

```python
df.columns = df.columns.str.replace('d_reg','reg_date')
df.columns
```
